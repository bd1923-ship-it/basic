```python
import pandas as pd
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = movies.merge(credits, left_on='id',right_on='movie_id')
movies.head()
```

```python
import ast
def convert(text):
   L=[]
   if isinstance(text, list):
     return text
   for i in ast.literal_eval(text):
    L.append(i[ 'name' ])
   return L
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
def fetch_cast(text):
    L=[]
    c=0
    if isinstance(text, list):
      return text[:3]
    for i in ast.literal_eval(text):
      if c != 3:
        L.append(i[ 'name' ])
        c+=1
      else:
        break
    return L
movies['cast'] = movies['cast'].apply(fetch_cast)
def fetch_director(text):
     if isinstance(text, str):
       try:
         data = ast.literal_eval(text)
         for i in data:
            if i['job'] == 'Director':
              return i['name']
       except (ValueError, SyntaxError):
         # If literal_eval fails, assume the text is already the director's name
         return text
     return []
movies['crew'] = movies['crew'].apply(fetch_director)
movies = movies[['movie_id','title_x','overview','genres','keywords','cast','crew']]
movies.loc[:, 'overview'] = movies['overview'].fillna('')
movies.head()
```

```python
movies['tags'] = movies['overview'].astype(str) + " " + movies['genres'].astype(str) + " " + movies['keywords'].astype(str) + " " + movies['cast'].astype(str) + " " + movies['crew'].astype(str)
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else str(x))
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer(max_features = 5000, stop_words='english')
vectors = cv.fit_transform(movies[ 'tags' ]).toarray()
similarity = cosine_similarity(vectors)

def recommend(movie):
  if movie not in movies[ 'title_x' ].values:
    return ["Movies not found!"]
  index = movies[movies[ 'title_x' ] == movie].index[0]
  distances = similarity[index]
  movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
  recommendations = []
  for i in movie_list:
    recommendations.append(movies.iloc[i[0]].title_x)
  return recommendations
```

```python

print("Recommendations for Avatar")
print(recommend("Avatar"))
print("\nRecommendation for Dark Knight")
print(recommend("The Dark Knight"))
```
